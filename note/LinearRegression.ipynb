```python
import tensorflow as tf
import numpy as np
import csv
import pandas as pd
```

```python
train_df = pd.read_csv("../input/house_train.csv")
train_df.head()
```

```python
test_df = pd.read_csv("../input/house_test.csv")
test_df.head()
```

```python
x_train = train_df.iloc[:, 0:13]
y_train = train_df[["MEDV"]]
x_test = test_df.iloc[:, 0:13]
y_test = test_df[["MEDV"]]
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x = tf.placeholder(tf.float32, [None, 13])
y = tf.placeholder(tf.float32, [None, 1])
W = tf.Variable(tf.zeros([13,1]))
b = tf.Variable(tf.zeros([1]))
tf.get_variable_scope().reuse_variables()
```

```python
y_reg = tf.matmul(x, W) + b
cost = tf.losses.mean_squared_error(labels = y, predictions = y_reg)
optimizer = tf.train.GradientDescentOptimizer(0.000001).minimize(cost)
```

```python
sess = tf.Session()
init = tf.initialize_all_variables()
sess.run(init)
num_epochs = 50000
for i in range(num_epochs):
    sess.run(optimizer, feed_dict = {x:x_train, y:y_train})
print(sess.run(W), sess.run(b))
```

```python
pred = sess.run(y_reg, feed_dict = {x: x_test})
rmse = ((pred - y_test) ** 2).mean() ** .5
print(rmse)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_norm = scaler.fit_transform(x_train)
x_test_norm = scaler.fit_transform(x_test)
```

```python
x = tf.placeholder(tf.float32, [None, 13])
y = tf.placeholder(tf.float32, [None, 1])
W = tf.Variable(tf.zeros([13,1]))
b = tf.Variable(tf.zeros([1]))
tf.get_variable_scope().reuse_variables()

y_reg = tf.matmul(x, W) + b
cost = tf.losses.mean_squared_error(labels = y, predictions = y_reg)
optimizer = tf.train.GradientDescentOptimizer(0.1).minimize(cost)
```

```python
sess = tf.Session()
init = tf.initialize_all_variables()
sess.run(init)
num_epochs = 500
for i in range(num_epochs):
    sess.run(optimizer, feed_dict = {x:x_train_norm, y:y_train})
print(sess.run(W), sess.run(b))
```

```python
pred = sess.run(y_reg, feed_dict = {x: x_test_norm})
rmse = ((pred - y_test) ** 2).mean() ** .5
print(rmse)
```
